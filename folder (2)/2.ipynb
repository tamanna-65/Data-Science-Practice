```python
def find_grade(marks):
  if (marks>=80):
    print('A+')
  elif (marks <80 and marks>=70):
    print('A')
  else:
    print('below A')

student_marks = [90,80,88,70,71,63,99]
for item in student_marks:
  find_grade(item)
```

```python
def calculate_price(quantity,unitprice):
  if (quantity > 1000):
    price = quantity*unitprice*.9
    print('price is ',price)
    return price

  else:
    price = quantity*unitprice
    print('price is ',price)
    return price

x = calculate_price(1001,100)

total = x+50
print(total)
```
