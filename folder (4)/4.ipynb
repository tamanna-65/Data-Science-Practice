```python
import pandas as pd
weight_data = [['tom',3.0],['pumpkin',3.51],['bonk',4.20],['thunder',5.50],['oreo',4.73],['nya',5.00],['kitkat',4.55],['bubbles',4.90],['sparkle',6.29],['pebbles',3.72]]
df_weight = pd.DataFrame(weight_data,columns = ['Name','Weight(kg)'] )
df_weight
```

```python
import seaborn as sns
sns.histplot(data = df_weight, x='Weight(kg)', bins=5)
```

```python
sns.FacetGrid(df_weight, height=6).map(sns.kdeplot, 'Weight(kg)').add_legend()
```

```python
import numpy as np
mean = np.mean(df_weight['Weight(kg)'])# mean<median(negatively skewed), mean>median(positively skewed)
mean
```

```python
median = np.median(df_weight['Weight(kg)'])
median
```

```python
weight_data = [['tom',0.3],['pumpkin',3.51],['bonk',4.20],['thunder',5.50],['oreo',4.73],['nya',5.00],['kitkat',4.55],['bubbles',4.90],['sparkle',6.29],['pebbles',3.72]]
df_weight = pd.DataFrame(weight_data,columns = ['Name','Weight(kg)'] )
df_weight
```

```python
sns.boxplot(y='Weight(kg)',data=df_weight)
```

```python
Q1 = np.percentile(df_weight['Weight(kg)'],25)
Q3 = np.percentile(df_weight['Weight(kg)'],75)

IQR = Q3-Q1

low= Q1-1.5*IQR
up = Q3+1.5*IQR

outlier=[]

for x in df_weight['Weight(kg)']:
  if((x>up) or (x<low)):
    outlier.append(x)
print('outliers are:',outlier)    
```
