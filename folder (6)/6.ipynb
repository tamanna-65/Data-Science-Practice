```python
data = pd.read_csv("OUTLIERS (2).csv")
data
```

```python
import numpy as np

q1= np.percentile(data['Sales'],25)
q3= np.percentile(data['Sales'],75)

iqr=q3-q1

upper = q3+1.5*iqr
lower = q1-1.5*iqr

outlier = []

for x in data["Sales"]:
  if (x > upper or x < lower):
    outlier.append(x)
  
  print(outlier)
```
